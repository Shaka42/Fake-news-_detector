```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix ,classification_report
```

```python
df = pd.read_csv('fakenews.csv')
df.head()
```

```python

df['label'] = df['label'].map({'FAKE':1,'REAL':0})
df
```

```python
features = df['title'].values
target = df['label'].values
```

```python
X_train, X_test, Y_train ,Y_test = train_test_split(features,target,test_size=0.2,random_state=65)
vectorizer = TfidfVectorizer(stop_words="english")
X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)
```

```python
results ={'RandomForestClassifier':RandomForestClassifier(n_estimators=7,random_state=0),'LogisticRegression':LogisticRegression(solver='liblinear',max_iter=100),'KNN classifier':KNeighborsClassifier(n_neighbors=2)}
for key ,value in results.items():
    a = value
    a.fit(X_train_bow,Y_train)
    y_predict = a.predict(X_test_bow)
    print(key,"\n")
    print(confusion_matrix(Y_test,y_predict),"\n")
    print(classification_report(Y_test,y_predict),"\n")
    print("The model score on test data:",a.score(X_test_bow,Y_test),"\n\n")
    results[key]= a.score(X_train_bow,Y_train)
print(results)
    
```
